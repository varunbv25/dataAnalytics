```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('marketing_campaign.csv',sep='\t')
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df.isnull())
```

```python
print(df.isnull().any(axis=1))
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna(axis=1)
```

```python
print(df_cleaned.shape)
print(df_cleaned.columns)
```

```python
df_no_duplicates=df_cleaned.drop_duplicates()
print(df_no_duplicates.shape)
df_no_duplicates.head()
```

```python
column_mapping = {
    'ID': 'id',
    'Year_Birth': 'year_birth',
    'Education': 'education',
    'Marital_Status': 'marital_status',
    'Income': 'income',
    'Kidhome': 'kids_home',
    'Teenhome': 'teens_home',
    'Dt_Customer': 'customer_date',
    'Recency': 'recency',
    'MntWines': 'amount_wines',
    'MntFruits': 'amount_fruits',
    'MntMeatProducts': 'amount_meat',
    'MntFishProducts': 'amount_fish',
    'MntSweetProducts': 'amount_sweets',
    'MntGoldProds': 'amount_gold',
    'NumDealsPurchases': 'num_deals_purchases',
    'NumWebPurchases': 'num_web_purchases',
    'NumCatalogPurchases': 'num_catalog_purchases',
    'NumStorePurchases': 'num_store_purchases',
    'NumWebVisitsMonth': 'num_web_visits_month',
    'AcceptedCmp3': 'accepted_campaign_3',
    'AcceptedCmp4': 'accepted_campaign_4',
    'AcceptedCmp5': 'accepted_campaign_5',
    'AcceptedCmp1': 'accepted_campaign_1',
    'AcceptedCmp2': 'accepted_campaign_2',
    'Complain': 'complain',
    'Z_CostContact': 'cost_contact',
    'Z_Revenue': 'revenue',
    'Response': 'response'
}

df_no_duplicates = df_no_duplicates.rename(columns=column_mapping)
```

```python
text_columns = df_no_duplicates.select_dtypes(include=['object']).columns
for col in text_columns:
    df_no_duplicates[col] = df_no_duplicates[col].str.strip()
```

```python
# Standardize Education
education_mapping = {
    'Graduation': 'Graduation',
    'PhD': 'PhD',
    'Master': 'Masters',
    'Basic': 'Basic',
    '2n Cycle': 'High School' # Assuming '2n Cycle' refers to high school level
}
df_no_duplicates['education'] = df_no_duplicates['education'].map(education_mapping)

# Standardize Marital Status
marital_mapping = {
    'Single': 'Single',
    'Alone': 'Single',
    'Together': 'In Relationship',
    'Married': 'Married',
    'Divorced': 'Divorced',
    'Widow': 'Widowed',
    'Absurd': 'Unknown', # Assuming 'Absurd' is an invalid entry
    'YOLO': 'Unknown' # Assuming 'YOLO' is an invalid entry
}
df_no_duplicates['marital_status'] = df_no_duplicates['marital_status'].map(marital_mapping)
```

```python
df_no_duplicates['customer_date'] = pd.to_datetime(df_no_duplicates['customer_date'], format='%d-%m-%Y', errors='coerce')
```

```python
integer_columns = [
    'id', 'year_birth', 'kids_home', 'teens_home', 
    'recency', 'num_deals_purchases', 'num_web_purchases',
    'num_catalog_purchases', 'num_store_purchases', 'num_web_visits_month',
    'accepted_campaign_1', 'accepted_campaign_2', 'accepted_campaign_3',
    'accepted_campaign_4', 'accepted_campaign_5', 'complain', 
    'cost_contact', 'response'
]

for col in integer_columns:
    df_no_duplicates[col] = pd.to_numeric(df_no_duplicates[col], errors='coerce').fillna(0).astype(int)
```

```python
float_columns = [
    'amount_wines', 'amount_fruits', 'amount_meat',
    'amount_fish', 'amount_sweets', 'amount_gold', 'revenue'
]

for col in float_columns:
    df_no_duplicates[col] = pd.to_numeric(df_no_duplicates[col], errors='coerce')
```

```python
df_no_duplicates.to_excel('marketing_campaign_cleaned.xlsx', index=False)
```
